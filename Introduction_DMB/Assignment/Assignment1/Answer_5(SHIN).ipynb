```python
import numpy as np
```

```python
def normalize_rows (matrix):

    return np.apply_along_axis(lambda row: row/np.sum(row) if np.sum(row) != 0 else row, axis =1, arr = matrix)
```

```python
matrix = np.array([[1, 2, 3, 4],
                  [2, 4, 6, 8],
                  [1, 3 ,5, 7]])
```

```python
result = normalize_rows(matrix)
result
```
