```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10 , 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
fig, ax = plt.subplots(1)

x = np.linspace(0, 10, 100)

ax.plot(x, np.sin(x));
```

```python
x = np.array([0, 10])
plt.plot(x, x, linestyle = '-')
plt.plot(x, x+1, linestyle = '--' )
plt.plot(x, x+2, linestyle = '-.')
```

```python
plt.plot(x, np.sin(x), 'k')
plt.plot(x, np.cos(x), '--r')
```

```python
x = np.array([0, 10])
plt.plot(x, x, linestyle = '-', lw=2)
plt.plot(x, x+1, linestyle = '--' ,lw=3)
plt.plot(x, x+2, linestyle = '-.', lw=4)
```

```python
x = np.arange(0, 10, 1)
plt.plot(x, x, 'o')
plt.plot(x, x+1, '-o')
plt.plot(x, x+2, 'k+')
plt.plot(x, x+3, '--s')
```

```python
x = np.linspace(0.1, 1, 10)
y1 = 1 / x
y2 = 10 * np.exp(-2* x)

plt.plot(x, y1, 'r--o', x, y2, 'k-.+')
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), label = 'sin(x)')
plt.plot(x, np.cos(x), label = 'cos(x)')


plt.legend(loc = 'upper center');
```

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x))

plt.title('A sine curve', fontsize = 18, fontname= 'Calibri')
plt.xlabel('x', fontsize =14, color='b')
plt.ylabel('sin(x)', fontsize = 14, color ='b');
```

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x))

plt.grid();
```

```python
x =np.linspace(-10, 10, 1000)
for n in range(1,5):
    y = x**n * np.sin(x)
    y  /=  max(y)
    plt.plot(x, y, label =fr'$x^{n}\sin x$')
plt.legend(loc = 'lower center');
```
